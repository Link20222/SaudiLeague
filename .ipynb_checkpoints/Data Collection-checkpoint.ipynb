```python
matches = data["api"]["fixtures"]
for i in range(len(matches)):
    
    fixture_id = matches[i]["fixture_id"]
    home_team = matches[i]["homeTeam"]["team_name"]
    away_team = matches[i]["awayTeam"]["team_name"]
    referee = matches[i]["referee"]
    nRound = matches[i]["round"].split("-")[1].replace(" ", "")
    date = matches[i]["event_date"]
    venue = matches[i]["venue"]
    home_goals = matches[i]["goalsHomeTeam"]
    away_goals = matches[i]["goalsAwayTeam"]
    status = matches["status"]
    
    season2018.loc[i] = [fixture_id, home_team,
                         away_team, home_goals,
                         away_goals, venue,
                         referee, nRound,
                         date, "2018/2019",
                         status]
```

### Developing a method to update all the matches registerd as not started to finished, by grapping all updated details

```python
last_finished_match = season2019.loc[season2019["status"] == "Match Finished"].tail(1).index

not_started_matches = season2019.loc[range(last_finished_match[0], len(season2019)), "fixture_id"]


url = "https://api-football-v1.p.rapidapi.com/v2/fixtures/league/901?timezone=Asia/Riyadh"

headers = {
    'x-rapidapi-host': "api-football-v1.p.rapidapi.com",
    'x-rapidapi-key': "d110139f75msh29b250f2354d68cp1f487ajsn3748c0ad1dca"
    }

response = requests.request("GET", url, headers=headers)

all_matches = response.json()

for i in not_started_matches.index:
    match = all_matches[i]
    
    if match["status"] == "Match Finished":
        fixture_id = match["fixture_id"]
        home_team = match["homeTeam"]["team_name"]
        away_team = match["awayTeam"]["team_name"]
        referee = match["referee"]
        nRound = match["round"].split("-")[1].replace(" ", "")
        date = match["event_date"]
        venue = match["venue"]
        home_goals = match["goalsHomeTeam"]
        away_goals = match["goalsAwayTeam"]
        status = match["status"]

        season2019.loc[i, ['fixture_id', 'home_team', 'away_team', 'home_goals', 'away_goals',
           'venue', 'referee', 'round','date' ,
                           'season', 'status']] = [fixture_id, home_team,
                                                     away_team, home_goals,
                                                     away_goals, venue,
                                                     referee, nRound,
                                                     date, "2019/2020",
                                   
                                                   status]

        url = f"https://api-football-v1.p.rapidapi.com/v2/lineups/{fixture_id}"

        headers = {
            'x-rapidapi-host': "api-football-v1.p.rapidapi.com",
            'x-rapidapi-key': "d110139f75msh29b250f2354d68cp1f487ajsn3748c0ad1dca"
            }

        response = requests.request("GET", url, headers=headers).json()
        match_details = response["api"]["lineUps"]
        #home team data

        home_coach = match_details[home_team]["coach"]
        home_formation = match_details[home_team]["formation"]

        home_players = []
        for player in match_details[match[home_team]]["startXI"]:
            
            home_players.append((player["player"], player["pos"]))


        #away team data                                             
        away_coach = match_details[match[away_team]["coach"]
        away_formation = match_details[match[away_team]]["formation"]

        away_players = []
        for player in match_details[match[away_team]]["startXI"]:
            
            away_players.append((player[away_team], player["pos"]))                                            

        season2019.loc[i, ["home_coach", "away_coach", "home_formation",
                           "away_formation", "home_players",
                           "away_players"]] = [home_coach, away_coach, home_formation,
                           away_formation, home_players, away_players]

```
