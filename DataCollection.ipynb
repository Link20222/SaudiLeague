```python
season2018 = pd.DataFrame(columns=["fixture_id", "home_team", "away_team", "home_goals", "away_goals",
                                        "venue", "referee","date", "season"])
```

```python
matches = data["api"]["fixtures"]
for i in range(len(matches)):
    
    fixture_id = matches[i]["fixture_id"]
    home_team = matches[i]["homeTeam"]["team_name"]
    away_team = matches[i]["awayTeam"]["team_name"]
    referee = matches[i]["referee"]
    nRound = matches[i]["round"].split("-")[1].replace(" ", "")
    date = matches[i]["event_date"]
    venue = matches[i]["venue"]
    home_goals = matches[i]["goalsHomeTeam"]
    away_goals = matches[i]["goalsAwayTeam"]
    
    season2018.loc[i] = [fixture_id, home_team, away_team, home_goals, away_goals, venue, referee, date, "2018/2019"]
```
